```python
import pandas as pd
```

```python
dataset = pd.read_csv("Marina_Bay_Sands.csv")
dataset.head()
```

```python
dataset['Date'][0][-9:]
```

```python
# by this I get 6 Dec from w 6 Dec 

for i in range(2,len(dataset['Date'])):
    if len(dataset['Date'][i].split()) > 2:
        dataset['Date'][i] = dataset['Date'][i][-8:]
```

```python
dataset['Date'][0]
```

```python
dataset.head()
```

```python
# by this if len(dataset['Date'].split) > 2 then Dec 2020

for i in range(len(dataset['Date'])):
    if len(dataset['Date'][i].split()) == 3:
        data = dataset['Date'][i].split()
        dataset['Date'][i] = data[1] +" "+data[2]
```

```python
dataset.head()
```

```python
# in it I convert yesterday into date

for i in range(len(dataset['Date'])):
    if dataset['Date'][i] == 'review Yesterday':
        dataset['Date'][i] = '18 Dec'
    
```

```python
dataset.head()
```

```python
dataset['Date'][1] = '17 Dec'
```

```python
dataset.head(16)
```

```python
# first 15 values in date column is like 1 Dec. So, I convert it into Dec 2020

for i in range(15):
    data = dataset['Date'][i][2:]
    dataset['Date'][i] = data+" 2020"
```
